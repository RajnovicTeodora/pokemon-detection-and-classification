```python
from numpy import expand_dims
from keras.preprocessing.image import load_img
from keras.preprocessing.image import img_to_array
from keras.preprocessing.image import ImageDataGenerator
from matplotlib import pyplot
from PIL import Image
import tensorflow as tf
import cv2
```

```python
def blur(img):
    return (cv2.blur(img,(5,5)))
```

```python
# create image data augmentation generator
datagen = ImageDataGenerator(preprocessing_function= blur)

def width_shift(samples, name, className):
    datagen = ImageDataGenerator(preprocessing_function=blur, width_shift_range=0.2)

    # prepare iterator
    it = datagen.flow(samples, batch_size=1)

    # generate samples and plot
    for i in range(9):
        # define subplot
        pyplot.subplot(330 + 1 + i)
        # generate batch of images
        batch = it.next()
        # convert to unsigned integers for viewing
        image = Image.fromarray(batch[0].astype('uint8'))
        # plot raw pixel data
        #pyplot.imshow(image)
        image.save('C:\\Users\\rajno\\Desktop\\pokemon-detection-and-classification\\data\\train\\' + className + '\\' + name + 'ws' + str(i) +'.jpg')
    # show the figure
    #pyplot.show()
```

```python
# example of vertical shift image augmentation
def height_shift(samples, name, className):
    # create image data augmentation generator
    datagen = ImageDataGenerator(preprocessing_function=blur, height_shift_range=0.1)

    # prepare iterator
    it = datagen.flow(samples, batch_size=1)

    # generate samples and plot
    for i in range(9):
        # define subplot
        pyplot.subplot(330 + 1 + i)
        # generate batch of images
        batch = it.next()
        # convert to unsigned integers for viewing
        image = Image.fromarray(batch[0].astype('uint8'))
        #plot raw pixel data
        #pyplot.imshow(image)
        image.save('C:\\Users\\rajno\\Desktop\\pokemon-detection-and-classification\\data\\train\\' + className + '\\' + name + 'hs' + str(i) +'.jpg')
    # show the figure
    #pyplot.show()
```

```python
# example of random rotation image augmentation

def rotate(samples, name, className):
    # create image data augmentation generator
    datagen = ImageDataGenerator(rotation_range=360)

    # prepare iterator
    it = datagen.flow(samples, batch_size=1)
    # generate samples and plot
    for i in range(9):
        # define subplot
        pyplot.subplot(330 + 1 + i)
        # generate batch of images
        batch = it.next()
        # convert to unsigned integers for viewing
        image = Image.fromarray(batch[0].astype('uint8'))
        # plot raw pixel data
        #pyplot.imshow(image)
        image.save('C:\\Users\\rajno\\Desktop\\pokemon-detection-and-classification\\data\\train\\' + className + '\\' + name + 'r' + str(i) +'.jpg')
    # show the figure
    #pyplot.show()
```

```python
# example of brighting image augmentation
def augment_brightness(samples, name, className):
    # create image data augmentation generator
    datagen = ImageDataGenerator(brightness_range=[0.2,1.5])

    # prepare iterator
    it = datagen.flow(samples, batch_size=1)
    # generate samples and plot
    for i in range(9):
        # define subplot
        pyplot.subplot(330 + 1 + i)
        # generate batch of images
        batch = it.next()
        # convert to unsigned integers for viewing
        image = Image.fromarray(batch[0].astype('uint8'))
        # plot raw pixel data
        #pyplot.imshow(image)
        image.save('C:\\Users\\rajno\\Desktop\\pokemon-detection-and-classification\\data\\train\\' + className + '\\' + name + 'ab' + str(i) +'.jpg')
    # show the figure
    #pyplot.show()
```

```python
# example of zoom image augmentation
def augment_zoom(samples, name, className):
    # create image data augmentation generator
    datagen = ImageDataGenerator(zoom_range=[0.8,0.8], rotation_range=360, shear_range=0.15,)

    # prepare iterator
    it = datagen.flow(samples, batch_size=1)

    # generate samples and plot
    for i in range(9):
        # define subplot
        pyplot.subplot(330 + 1 + i)
        # generate batch of images
        batch = it.next()
        # convert to unsigned integers for viewing
        image = Image.fromarray(batch[0].astype('uint8'))
        # plot raw pixel data
        #pyplot.imshow(image)
        image.save('C:\\Users\\rajno\\Desktop\\pokemon-detection-and-classification\\data\\train\\' + className + '\\' + name + 'az' + str(i) +'.jpg')
    # show the figure
    #pyplot.show()
```

```python
# load the image

import os
os.getcwd()
classes = ['Abra', 'Arcanine', 'Bulbasaur','Ditto', 'Eevee', 'Gastly', 'Gengar', 'Jigglypuff', 'Pikachu']

for name in classes:
    collection = "C://Users//rajno//Desktop//pokemon-detection-and-classification//data//train//" + name
    for i, filename in enumerate(os.listdir(collection)):
        #print(filename)
        img = load_img('data/' + name + '/' + filename)
        
        # convert to numpy array
        data = img_to_array(img)
        
        # expand dimension to one sample
        samples = expand_dims(data, 0)
        
        width_shift(samples, filename.split('.')[0], name)
        height_shift(samples, filename.split('.')[0], name)
        expand(samples, filename.split('.')[0], name)
        rotate(samples, filename.split('.')[0], name)
        augment_brightness(samples, filename.split('.')[0], name)
        augment_zoom(samples, filename.split('.')[0], name)
```
